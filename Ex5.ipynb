```python
import psycopg2
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
hostname = 'localhost'
database = 'postgres'
username = 'postgres'
# pws = '123456'
pws = 'postgres'
port_id = 5432

conn = None
cur = None
```

```python
conn = psycopg2.connect(
    host = hostname,
    dbname = database,
    user = username,
    password = pws,
    port = port_id
)


cur = conn.cursor(cur)
```

```python

cur.execute("SELECT D.Dnome AS NomeDepartamento, COUNT(F.Cpf) AS NumFuncionarios \
            FROM DEPARTAMENTO D \
            JOIN FUNCIONARIO F ON D.Dnumero = F.Dnr \
            GROUP BY D.Dnome \
            HAVING AVG(F.Salario) > 30000.00;")
data = cur.fetchall()
df = pd.DataFrame(data)
df
```

```python
cur.execute("SELECT D.Dnome AS NomeDepartamento, COUNT(F.Cpf) AS NumFuncionariosMasculinos \
            FROM DEPARTAMENTO D \
            JOIN FUNCIONARIO F ON D.Dnumero = F.Dnr \
            WHERE F.Sexo = 'M' AND F.Salario > 30000.00 \
            GROUP BY D.Dnome;")
data = cur.fetchall()
df = pd.DataFrame(data)
df
```

```python
cur.execute("SELECT * FROM FUNCIONARIO;")
data = cur.fetchall()
df = pd.DataFrame(data)
df
```

```python
cur.execute("SELECT F.Pnome || ' ' || F.Unome AS NomeFuncionario \
            FROM FUNCIONARIO F \
            WHERE F.Dnr = ( \
                SELECT D.Dnumero \
                FROM DEPARTAMENTO D \
                JOIN FUNCIONARIO F2 ON D.Dnumero = F2.Dnr \
                WHERE F2.Salario = ( \
                    SELECT MAX(Salario) \
                    FROM FUNCIONARIO));")
data = cur.fetchall()
df = pd.DataFrame(data)
df
```
